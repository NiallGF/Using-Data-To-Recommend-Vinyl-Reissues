```python
import pandas as pd
import matplotlib.pyplot as plt

# Data creation
data = {
    "Month": ['January', 'January', 'January', 'February', 'February', 'February', 'March', 'March', 'March'],
    "Product": ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    "Sales": [345, 750, 150, 140, 455, 600, 300, 150, 145],
    "AdvertisingSpend": [100, 130, 80, 70, 100, 110, 90, 60, 80]
}

%matplotlib inline
rcParams['figure.figsize'] = 12, 7

df = pd.DataFrame(data)

# Filter for Product A
product_a = df[df['Product'] == 'A']

# Convert 'Month' to datetime format for better x-axis handling
product_a['Month'] = pd.to_datetime(product_a['Month'] + ' 2024', format='%B %Y')

# Create the plot
plt.plot(product_a['Month'], product_a['Sales'], marker='o', label='Product A Sales')
plt.title('Sales Trend for Product A')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_total_sales = df.groupby("Product")["Sales"].sum()
df_total_sales

plt.bar(df_total_sales.index, df_total_sales.values)
plt.title("Total Sales Of Each Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
```

```python
plt.scatter(df["AdvertisingSpend"],df["Sales"],color='r', label='Individual Sales')
plt.title("Advertising Budget vs. Sales")
plt.xlabel("Advertising Budget")
plt.ylabel("Sales")
plt.legend()
```
