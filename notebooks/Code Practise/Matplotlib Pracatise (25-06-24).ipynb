```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams

data = {
    "Year": [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    "GlobalTempChange": [0.42, 0.54, 0.63, 0.62, 0.54, 0.67, 0.64, 0.66, 0.54, 0.64, 0.72, 0.60, 0.65, 0.67, 0.74, 0.90, 0.99, 0.91, 0.85, 0.95, 1.02],
    "CO2Emissions": [29.2, 29.9, 30.6, 31.3, 31.9, 32.6, 33.4, 34.2, 34.9, 35.5, 36.1, 36.7, 37.4, 38.1, 38.8, 39.5, 40.1, 40.8, 41.5, 42.2, 42.9]
}

df = pd.DataFrame(data)
df
```

**Visualize the Trends:** Create a line plot to visualize the trends of global temperature change and CO2 emissions over the years.

```python
%matplotlib inline
rcParams['figure.figsize'] = 12, 7

plt.plot(df["Year"], df["CO2Emissions"], color='r', label='CO2 Emissions')

plt.plot(df["Year"],df["GlobalTempChange"], color = 'b', label = 'Global Temperature')

plt.title('Global Temperature Change and CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid()

plt.show()
```

**Correlation Analysis:** Calculate and display the correlation between global temperature change and CO2 emissions.

```python
correlation = df["GlobalTempChange"].corr(df["CO2Emissions"])
correlation
```

```python
df['5-Year Moving Avg'] = df['GlobalTempChange'].rolling(window=5).mean()

```

```python
# Set figure size
plt.figure(figsize=(12, 7))

# Plot original Global Temperature Change
plt.plot(df['Year'], df['GlobalTempChange'], color='blue', label='Global Temperature Change')

# Plot 5-Year Moving Average
plt.plot(df['Year'], df['5-Year Moving Avg'], color='red', linestyle='--', label='5-Year Moving Avg')

# Titles and Labels
plt.title('Global Temperature Change with 5-Year Moving Average')
plt.xlabel('Year')
plt.ylabel('Global Temperature Change (°C)')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
