```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

**Concat By Column Of Above Tables** (Note that the concat of columns is done by 'axis = 1')

```python
pd.concat([DF_obj, DF_obj_2], axis=1)
```

**Concat By Rows Of Above Tables** (Note that the concat of rows does not need to have an axis command)

```python
pd.concat([DF_obj, DF_obj_2])
```

**Below Command To Drop Rows/Columns** (Again, if 'axis = 1' was used then the columns would be removed)

```python
DF_obj.drop([0,2])
```

**Below Table To Be Used In Concat**

```python
series_obj = Series(np.arange(6))
series_obj.name = "added_variable"
series_obj
```

**Below Joins DF_obj & series_obj** (Must do more research on when join or concat is best to be used)

```python
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

**Below Concatenates DF_obj & series_obj** 
- Must research when join or concat is best to be used.
- Also - 'ignore index = True' will ignore the original index (not how it is 0-11). If false, the index will repeat as it will show the original indexes joined together

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index=True)
added_datatable
```

**Sorting By Row/Column** 
- Same applies as before - can select row or column by using axis.
- Additionally, by= (5) is what is selecting the row and can select ascending or not.

```python
DF_sorted = DF_obj.sort_values(by=(5), ascending = False)
DF_sorted
```
