```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

**Missing Data**

```python
missing = np.nan

series_obj = Series(['row 1', 'row 2', missing, 'row 4', 'row 5', 'row 6', missing, 'row 8'])
series_obj
```

**Boolean Of Missing Values**

```python
series_obj.isnull()
```

**Filling In Missing Values**

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6, 6))
DF_obj
```

```python
DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

```python
filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
filled_DF = DF_obj.fillna({0:0.4, 5:1.25})
filled_DF
```

**Fill Forward** (fills empty with preceding val's)

```python
fill_DF = DF_obj.fillna(method='ffill')
fill_DF
```

**Count Missing Values**

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6, 6))
DF_obj

DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

**Count Null Cells By Column**

```python
DF_obj.isnull().sum()
```

**Drop All Rows Containing NaN (Not A Number/Null)**

```python
DF_no_NaN  = DF_obj.dropna()
DF_no_NaN
```

**Drop All Columns Containing NaN (Not A Number/Null)**

```python
DF_no_NaN  = DF_obj.dropna(axis=1)
DF_no_NaN
```
